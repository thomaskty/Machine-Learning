```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the range of x values
x_values = np.linspace(0.1, 10, 100)

# Calculate the corresponding y values for log base 2, 3, 10, and natural log (loge)
y_log2 = np.log2(x_values)
y_log3 = np.log2(x_values) / np.log2(3)  # log base 3 = log base 2 / log base 2 of 3
y_log10 = np.log10(x_values)
y_loge = np.log(x_values)

# Plotting the functions
plt.figure(figsize=(10, 3))
plt.plot(x_values, y_log2, label='log2(x)')
plt.plot(x_values, y_log3, label='log3(x)')
plt.plot(x_values, y_log10, label='log10(x)')
plt.plot(x_values, y_loge, label='loge(x)')
plt.xticks(list(range(10)))
# Adding labels and title
plt.xlabel('x')
plt.ylabel('log x')
plt.title('Logarithmic Functions')
plt.legend()

# Display the plot
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x_values = np.linspace(0.01, 10, 100)  # Start from a small value to avoid division by zero

# Calculate the corresponding y values for log base 2, 3, 10, and natural log (loge) of (1/x)
y_log2 = np.log2(1/x_values)
y_log3 = np.log2(1/x_values) / np.log2(3)
y_log10 = np.log10(1/x_values)
y_loge = np.log(1/x_values)

# Plotting the functions
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_log2, label='log2(1/x)')
plt.plot(x_values, y_log3, label='log3(1/x)')
plt.plot(x_values, y_log10, label='log10(1/x)')
plt.plot(x_values, y_loge, label='loge(1/x)')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Logarithmic Functions (1/x)')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Define the range of x values
x_values = np.linspace(0.1, 10, 100)

# Calculate the corresponding y values for log base 2, 3, 10, and natural log (loge)
y_log2 = x_values*np.log2(x_values)
y_log3 = x_values*np.log2(x_values) / np.log2(3)  # log base 3 = log base 2 / log base 2 of 3
y_log10 = x_values*np.log10(x_values)
y_loge = x_values*np.log(x_values)

# Plotting the functions
plt.figure(figsize=(10, 3))
plt.plot(x_values, y_log2, label='log2(x)')
plt.plot(x_values, y_log3, label='log3(x)')
plt.plot(x_values, y_log10, label='log10(x)')
plt.plot(x_values, y_loge, label='loge(x)')
plt.xticks(list(range(10)))
# Adding labels and title
plt.xlabel('x')
plt.ylabel('x log x')
plt.title('Graph of x log x')
plt.legend()

# Display the plot
plt.grid()
plt.show()
```
